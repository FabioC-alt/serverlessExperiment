```python
import pandas as pd
import matplotlib.pyplot as plt

# Load each file (single column, no header)
df1 = pd.read_csv('lateciesGO.txt', header=None, names=['y1'])

# Plot lines using the row index as x
plt.plot(df1.index, df1['y1'], label='Line 1')

plt.xlabel('Index (Sample Number)')
plt.ylabel('Value')
plt.title('Plot from Three Single-Column Files')
plt.legend()
plt.grid(True)
plt.show()
```
