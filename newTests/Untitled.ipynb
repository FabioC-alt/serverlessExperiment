```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess data
df = pd.read_csv('Kafka3Node.txt', delimiter='\t')
df['Values_ms'] = df['Values'] * 1000

# Compute 99th percentile
p99 = df['Values_ms'].quantile(0.99)

# Create a simple bar chart
plt.figure(figsize=(4, 5))
sns.barplot(x=['99th Percentile'], y=[p99], color='skyblue')

plt.ylabel('Latency (ms)')
plt.title('99th Percentile Latency: 3 Node, RabbitMQ')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
p50 = df['Values_ms'].quantile(0.50)
p95 = df['Values_ms'].quantile(0.95)

plt.figure(figsize=(6, 5))
sns.barplot(
    x=['50th Percentile', '95th Percentile', '99th Percentile'],
    y=[p50, p95, p99],
    palette='Blues_d'
)

plt.ylabel('Latency (ms)')
plt.title('Latency Percentiles: 3 Node, Kafka')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('kafka_99th_percentile_latency.png', dpi=300)
plt.show()


```
