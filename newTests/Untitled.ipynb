```python
import pandas as pd
df = pd.read_csv('latencies_3node.txt', delimiter='\t')
import matplotlib.pyplot as plt
import seaborn as sns

df['Values_ms'] = df['Values'] * 1000
print(df.head())


plt.figure(figsize=(8, 3))
sns.violinplot(x='Values_ms', data=df, inner='box', color='skyblue')

plt.xlabel('Latency (ms)')
plt.title('Horizontal Violin Plot of Latency Values: 3 Node, RabbitMQ')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
