```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('latency_js_minio_endLatency_100_FabioPc.csv',sep=';',index_col=0)
df.drop(columns=['Unnamed: 3'], inplace=True)
cols = df.columns
print(cols)
df.head()
```

```python
median_values = df.median()
print(f"Median Value in ns: {median_values.iloc[0]}")
median_second = median_values.iloc[0]/1e9
print(f"Median Value in s: {median_second}")
```

```python
df = df.drop(index=1)

df[' latency_ms'].plot(title='Latency in ms')
plt.xlabel('Iteration')
plt.ylabel('Latency (ms)')
plt.grid(True)
plt.show()
```
