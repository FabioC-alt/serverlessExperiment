```python
import pandas as pd
import matplotlib.pyplot as plt
import re

def extract_latencies(filename):
    latencies = []
    with open(filename, 'r') as f:
        for line in f:
            # Use regex to find the latency value before ' ms'
            match = re.search(r': ([\d.]+) ms', line)
            if match:
                latencies.append(float(match.group(1)))
    return latencies

# Load latencies from each file
latencies1 = extract_latencies('latenciesFN.txt')

# Create dataframes or just use lists with index as x
plt.plot(range(len(latencies1)), latencies1, label='File 1')

plt.xlabel('Request Number')
plt.ylabel('Latency (ms)')
plt.title('Latencies from Three Files')
plt.legend()
plt.grid(True)
plt.show()
```
