```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica e ordina i dati
df = pd.read_csv('latency.csv')
df = df.sort_values(by='request_number')

# Calcola i percentili
p25 = df['latency_ms'].quantile(0.25)
p50 = df['latency_ms'].quantile(0.50)
p75 = df['latency_ms'].quantile(0.75)

# Imposta dimensione figura più grande
plt.figure(figsize=(12, 7))

# Grafico latency vs request_number senza marker
plt.plot(df['request_number'], df['latency_ms'], linestyle='-', label='Latency')

# Linee percentile
plt.axhline(p25, color='green', linestyle='--', label='25th Percentile')
plt.axhline(p50, color='orange', linestyle='--', label='50th Percentile (Median)')
plt.axhline(p75, color='red', linestyle='--', label='75th Percentile')

plt.title('Latency per Request Number with Percentiles')
plt.xlabel('Request Number')
plt.ylabel('Latency (ms)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carica i dati da CSV
# Assumiamo che il file abbia una colonna "latency" e una colonna "timestamp" o indice temporale
df = pd.read_csv('latency.csv')

# Rimuovi outlier usando IQR (interquartile range)
Q1 = df['latency_ms'].quantile(0.25)
Q3 = df['latency_ms'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_clean = df[(df['latency_ms'] >= lower_bound) & (df['latency_ms'] <= upper_bound)]

# Applica media mobile con finestra di 10 campioni per smoothing
window_size = 10
df_clean['latency_smooth'] = df_clean['latency_ms'].rolling(window=window_size, min_periods=1).mean()

# Plot lineare
plt.figure(figsize=(12,6))
plt.plot(df_clean.index, df_clean['latency_ms'], label='Raw latency', alpha=0.3)
plt.plot(df_clean.index, df_clean['latency_smooth'], label='Smoothed latency', linewidth=2)
plt.title('Latency over time (Linear scale)')
plt.xlabel('Sample Index')
plt.ylabel('Latency (ms)')
plt.legend()
plt.grid(True)
plt.show()

# Plot logaritmico (evita valori <= 0, filtra o aggiungi epsilon)
epsilon = 1e-3
latency_log = df_clean['latency_smooth'].clip(lower=epsilon)
plt.figure(figsize=(12,6))
plt.plot(df_clean.index, latency_log)
plt.yscale('log')
plt.title('Latency over time (Logarithmic scale)')
plt.xlabel('Sample Index')
plt.ylabel('Latency (ms, log scale)')
plt.grid(True, which="both", ls="--")
plt.show()
```
