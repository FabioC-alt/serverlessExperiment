```python
import pandas as pd
import matplotlib.pyplot as plt

# Load each file (single column, no header)
df1 = pd.read_csv('lateciesGO.txt', header=None, names=['y1'])
df1 = df1.iloc[500:2500]

df2 = pd.read_csv('latenciesFN.txt', header=None, names=['y2'])
df2 = df2.iloc[500:2500]

df3 = pd.read_csv('latenciesJava.txt', header=None, names=['y3'])
df3 = df3.iloc[500:2500]

# Plot lines using the row index as x
plt.plot(df1.index, df1['y1'], label='Go')
plt.plot(df2.index, df2['y2'], label='Javascript')
plt.plot(df3.index, df3['y3'], label='Java')

# Set axis labels and title
plt.xlabel('Index (Sample Number)')
plt.ylabel('Latency')
plt.title('Plot from Three Single-Column Files')

# Set y-axis to log scale
plt.yscale('log')

# Add legend and grid
plt.legend()
plt.grid(True)

# Save and show the plot
plt.savefig('latency_plot.png', dpi=300)
plt.show()
```
